```python
import pandas as pd
main_berlin_search = "https://www.wg-gesucht.de/wg-zimmer-in-Berlin.8.0.1." # 0.html

#add all berlin-wg-search pages into a list
search_page_list = []
for i in range(1766): #there were 1766 search pages for Berlin at the time
    search_page_url = main_berlin_search + str(i) + ".html"
    search_page_list.append(search_page_url)
search_page_series = pd.Series(search_page_list)

#a typical recorded url of search pages is like the following:
search_page_series[30]
print(len(search_page_series))
```

```python
#now, it is time to go through all of the search pages, look into their source codes and retrieve advertisement URLS

from urllib.request import urlopen
from bs4 import BeautifulSoup


ad_URL_list = []  #create a blank list for collecting ad URls later     
all_URL_list = [] #create a blank list for collecting all URls later  

for page in search_page_series[270:280]:
    print(page)
    html = urlopen(page) # Insert your URL to extract
    html_scrap_object = BeautifulSoup(html.read(),"lxml");

    #first we retrieve all .html links out of the search page
    for link in html_scrap_object.find_all('a'):
        ad_page_url = str(link.get('href'))
        if ad_page_url[-5:] == ".html":  #urls will be discarded unless they end with ".html"
            all_URL_list.append(ad_page_url)

#then we will create a list of all URLs linked in the page, that has the format .../xxxxxxx.html
#where xxxxxxx is a 7 digit integer; since all ad-pages have this format in the site 
for item in list(set(all_URL_list)):   #make sure to iterate over unique urls list, web pages include certain URLs more than once
    if any(char.isdigit() for char in item[-6]) & \
        any(char.isdigit() for char in item[-7]) & \
        any(char.isdigit() for char in item[-8]) & \
        any(char.isdigit() for char in item[-9]) & \
        any(char.isdigit() for char in item[-10]) & \
        any(char.isdigit() for char in item[-11]) & \
        any(char.isdigit() for char in item[-12]) == True:
        ad_URL_list.append(item)
        
print(list(set(ad_URL_list)))
```

```python
import csv
ad_url_list_final = []
with open('wg-gesucht-url-list.csv', 'r', encoding="utf8") as f:
    reader = csv.reader(f)
    for row in reader:
        ad_url_list_final.append(str(row).replace("[", "").replace("]", "").replace("'",""))

ad_page_url_series = pd.Series(ad_url_list_final)
```
